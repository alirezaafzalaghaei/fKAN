```python
import glob
import os
import re

import matplotlib.pyplot as plt
import numpy as np
import scienceplots
import scipy.stats as stats

plt.style.use("science")
```

```python
files = sorted(glob.glob("data/m-*.npy"))
fig = plt.figure(figsize=(8, 3.5))
line_styles = ["-", "--", "-.", ":", "solid", "dashed"]
for m, file in enumerate(files, 0):
    predict = np.load(file)[::100]
    x = np.linspace(0, 15, len(predict))
    if m == 1:
        x = x[:40]
        predict = predict[:40]
    plt.plot(x, predict, label="$m=%d$" % m, ls=line_styles[m])

plt.ylim(0, 1.1)
plt.xlim(0, 15)
plt.legend()

plt.xlabel(r"$\zeta$")
plt.ylabel(r"$\hat{\chi}(\zeta)$")

fig.savefig("lane-emden.pdf", bbox_inches="tight")
```
